Lo que recomienda el libro es definir un valor tomado como outlier y luego a este mimso se lo puede limitar a un valor aceptado o eliminar

```python
import numpy as np 
import pandas as pd
```

```python
data = pd.DataFrame(np.random.randn(1000, 4))
data
```

```python
data[(np.abs(data) > 3).any(axis=1)]
```

```python
data[np.abs(data) > 3] = np.sign(data) * 3
```

```python
data[(np.abs(data) > 3).any(axis=1)]
```

```python
data[(np.abs(data) == 3).any(axis=1)]
```
