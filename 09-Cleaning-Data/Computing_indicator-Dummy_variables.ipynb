```python
columns = ['id', 'title', 'genres']
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/amirtds/kaggle-netflix-tv-shows-and-movies/refs/heads/main/titles.csv',usecols=columns)
```

```python
df.head(5)
```

```python
df['genres'] = df['genres'].map(lambda x: x.replace('[','').replace("'","").replace(']','').replace(' ',''))
```

```python
df.head(10)
```

```python
all_genres = []
for x in df.genres:
    all_genres.extend(x.split(','))
all_genres
```

```python
genres = pd.Series(all_genres).unique()
genres
```

```python
zero_matrix = np.zeros((len(df), len(genres)))
dummies = pd.DataFrame(zero_matrix, columns=genres)
```

```python
gen = df.genres[0]
gen.split(',')
```

```python
dummies.columns
```

```python
dummies.columns.get_indexer(gen.split(','))
```

```python
for i, gen in enumerate(df.genres):
    indices = dummies.columns.get_indexer(gen.split(','))
    dummies.iloc[i, indices] = 1
```

```python
movies_windic = df.join(dummies.add_prefix('Genre_'))
```
